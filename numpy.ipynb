```python
y = np.array([1,2,3,4,5])
print(y.ndim) # dimension count
print(y.shape) # shape of array
print(y.size) # element count
```

```python
z = np.array([[1,2,3],[3,4,5],[5,6,7]])
print(z.ndim) # dimension count
print(z.shape) # shape of array
print(z.size) # element count
```

```python
a = np.arange(10,100,.1)
print(a.ndim) # dimension count
print(a.shape) # shape of array
print(a.size) # element count
```

```python
a2 = a.reshape((30,30))
print(a2.ndim) # dimension count
print(a2.shape) # shape of array
print(a2.size) # element count
```

```python
a3 = a.reshape((30,15,2))
print(a3.ndim) # dimension count
print(a3.shape) # shape of array
print(a3.size) # element count
```

```python
b = np.linspace(1,10,100)
print(b)
```

```python
a = np.random.randint(1,10,(3,3))
b = np.random.randint(1,10,(3,3))
print(a)
print(b)
print(a+b)
print(a-b)
print(a * b)
print(a / b)
print(a > b)
print(a < b)
```

```python
print(a.max())
print(a.mean())
print(a.min())
print(a.std())
print(a.sum())
```

```python
print(a.max(axis=0)) # row
print(a.max(axis=1)) # column
```

```python
t = np.arange(-np.pi * 3, np.pi, .05)
x = np.sin(t)
y = np.cos(t)
print(t)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(t,x,'r--o')
plt.plot(t,y,'g:x')
plt.title("Visualisation")
plt.show()
```
